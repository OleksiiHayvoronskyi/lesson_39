```python
from sklearn.datasets import load_iris
from sklearn import tree


iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.3, shuffle=True)
print('The model has been splitted!')
```

```python
from sklearn.metrics import accuracy_score

# Creating Decision Tree Classifier.
dtree_model = DecisionTreeClassifier(criterion='gini', random_state=1)
dtree_model.fit(X_train, y_train)

# Predict Accuracy Score.
y_pred = dtree_model.predict(X_test)

train_score = accuracy_score(y_true=y_train, y_pred=dtree_model.predict(X_train))
test_score = accuracy_score(y_true=y_test, y_pred=y_pred)

print(f"Train data accuracy: {train_score:.2f}" )
print(f"Test data accuracy: {test_score:.2f}")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,10))
get_fig = tree.plot_tree(dtree_model, feature_names=iris.feature_names, filled=True)
```
